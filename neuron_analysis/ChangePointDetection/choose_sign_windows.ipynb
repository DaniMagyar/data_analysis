```python
def check_results_by_user(result, f_result):
    print(f"The filtered change points are:{f_result}")
    next_point = result.index(f_result[-1])+1
    wantoadd = input(f"Would you like to add the following changepoint ({result[next_point]}) to the results? Type Yes or leave empty for no.")
    
    while wantoadd == "Yes" and f_result[-1] != result[-1]:
        f_result.append(result[next_point])
        check_results_by_user(result,f_result)
        return f_result
    return f_result
```

```python
def get_cell_ID(filename):
    filename= filename.replace("_", "")
    animal = re.search(r"(FP\d+)",filename).group(0)
    cell = re.search(r"(unit\d\d)",filename).group(0)
    return "_".join([animal,cell])
```

```python
# for individual files
def calc_change_points(evoked_spikes_file,N,min_T):
    multiplier = 1
    output = {}
    data = pd.read_csv(file)
    bins = dict.fromkeys(range(1,N+1), 0) # create dictionary with keys 0 to n_bins all with value 0
    for spike in data["Delay from last TTL"]:

        binvalue = math.ceil(spike / bin_size)
        
        
        if spike < min_T+0.01 and binvalue == (N+1):
            binvalue -= 1
        bins[int(binvalue)] += 1
    offset = round(N/3)
    x_vals = []
    y_vals = []
    for i in range(N):
        x_vals.append(i*bin_size)
        y_vals.append(bins[i+1])

    counter = -1
    for j in range(N,N-offset,-1):
        x_vals.insert(0,counter*bin_size)
        y_vals.insert(0,bins[j])
        counter += -1
    z_scores = stats.zscore(y_vals)    
    model="rank"
    std_noise = np.std(z_scores[:offset])
    std_noise_full =np.std(z_scores)
    print(std_noise_full)
    pen = np.log(N) * 1 * std_noise**2
    print(pen)
    print("INITIAL TIMEPOINT: ",(N-(N-offset)))

    algo = rpt.Pelt(model=model,jump=1).fit(np.array(z_scores))
    result = algo.predict(pen=pen)

    f_result = [x for x in result if ( N*multiplier > x > ((N-(N-offset)))*multiplier-2 )][:4]
    rpt.display(np.array(z_scores), result, figsize=(30, 15))
    plt.title(file)
    plt.xticks(np.arange(0, N*multiplier, step=10))
    plt.show()
    if len([x for x in result if N*multiplier > x> ((N-(N-offset))-2)*multiplier]) > 4:
        f_result = check_results_by_user(result=result,f_result=f_result)

    ID = get_cell_ID(os.path.basename(file))
    print(ID)
    print(result)
    print(f_result)
    output[ID] = [(x-(N-(N-offset)))*multiplier/10 for x in f_result]
    print(output)
    return output
```

```python
def convert_changepoints_to_segments(dictionary):
    # changepoints converted to segment ranges [from,to]
    # and save to DataFrame object
    df = pd.DataFrame()
    for key, value in dictionary.items():
        pairs = list(zip(value,value[1:]))
        sub_df = pd.DataFrame({key:pairs})
        df = pd.concat([df, sub_df], axis=1)
    return df
        
        
    
```

```python
def concat_datas(df_r, df_c):
    # concat read dataframe (df_r) and current dataframe (df_c)
    # firstly check whether columns from df_c exist in df_r
    # then add non-existing data
    non_existing_cols = []
    for col in df_c.columns:
        if not col in df_r.columns:
            df_r = pd.concat([df_r,df_c], axis=1)
            non_existing_cols.append(col)
    if len(non_existing_cols) > 0:
        print(f"The following datas has been added successfully:{non_existing_cols}.")
    else:
        print(f"No new data were found.The table has not been updated.")
    return df_r
    
def write_all_to_csv(target_path,current_df):
    if not os.path.exists(target_path):
        current_df.to_csv(target_path)
        print(f"The following file had been created: {target_path}")
        print(f"The following datas has been added successfully: {current_df.columns}")
        return current_df
    else:
        df_f = pd.read_csv(target_path, index_col=[0])
        df_f = concat_datas(df_f,current_df)
        df_f.to_csv(target_path)
        return df_f
```
